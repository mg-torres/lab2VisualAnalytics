 ### Limpieza de Terridata

```python
#Se eliminan los nulos
terridata = terridata.dropna(axis = "index", how = "any");
```

```python
terridata.notna()
```

```python
#Se eliminan los registros duplicados
terridata = terridata.drop_duplicates();
```

```python
#Se cambia el nombre de dato numérico por un nombre que sea más fácil de utilizar en Observable(en este caso Valor)
#Lo mismo se hace con Código Departamento, este se cambia por id

terridata.rename(columns = {'Dato Numérico':'Valor'}, inplace = True)
terridata.rename(columns = {'Código Departamento':'id'}, inplace = True)
```

## Tarea 1 

```python
terridata
```

```python
# Se filtran los datos para tener únicamente el inidicador de la tasa de repitencia

dftarea1 = terridata.copy()

dftarea1 = dftarea1[dftarea1.Indicador.str.contains('Tasa de repitencia')]
```

```python
tarea1 =  dftarea1.groupby(['Departamento','Entidad', 'Año','Indicador'])['Valor'].mean()
```

## Tarea 2


```python
## Se filtran los datos de la tarea 2 para tener en cuenta únicamente los items relacionados con el cubrimiento

dftarea2 = terridata.copy()

dftarea2 = dftarea2[dftarea2.Indicador.str.contains('Cobertura neta')]


dftarea2
```

```python
# Como se quiere analizar a toda la población, se eliminan los datos que contengan información sobre la cobertura neta en hombres o mujeres}

dftarea2 = dftarea2[dftarea2['Indicador'].str.contains(' - Mujeres| - Hombres')==False]

dftarea2.dtypes
```

```python
dftarea2
```

```python
# Se filtran los datos para obtener únicamente la cobertura neta en educación total para el departamento 

dftarea2 = dftarea2[dftarea2.Indicador.str.contains('Cobertura neta en educación - Total')]
```

```python
dftarea2
```

```python
#El id en el geojson de los códigos de un digito es "01", "02" etc. y en el csv no tiene el cero a la izquierda, por lo que se estandariza este id 
#colocándole un cero a la izquierda a los códigos de un dígito.

dftarea2['id'] = dftarea2['id'].apply(lambda x: x.zfill(2))
```

```python
#Se agrupan los datos por id, año, departamento e indicador y se calcula el promedio del cubrimiento de educación total en cada departamento(sacando el promedio de los municipios del departamento). 
tarea2 =  dftarea2.groupby(['id','Año','Departamento', 'Indicador'])['Valor'].mean()
```

```python
tarea2
```

```python
dftarea2
```

```python
#Para facilitar la implementación de la visualizaicón, se calcula el cubrimiento promedio, únicamente agrupando por año y por id.
tarea2simplificado= dftarea2.groupby(['Año', 'id'])['Valor'].mean()
```

```python
tarea2simplificado
```

```python
tarea2
```

# Tarea 3

```python
# Se filtra y se saca la media del cubrimiento promedio por departamento para la tarea 3 

dftarea3 = terridata.copy()

dftarea3 = dftarea3[dftarea3.Indicador.str.contains('Cobertura neta')]
dftarea3 = dftarea3[dftarea3['Indicador'].str.contains(' - Mujeres| - Hombres')==False]
dftarea3
```

```python
# Sacamos el promedio de el dato nummérico para poder graficar 
tarea3 =  pd.DataFrame(dftarea3[['Departamento','Entidad', 'Subcategoría', 'Año','Indicador', 'Valor', 'Unidad de Medida' ]].groupby(['Departamento', 'Subcategoría', 'Año', 'Indicador', 'Valor']).mean())
```

```python
# Agrupamos los datos por departamento y año y calculamos el cubrimiento promedio
tarea3 =  dftarea3.groupby(['Departamento','Año'])['Valor'].mean()
```
